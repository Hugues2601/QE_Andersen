```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Télécharger les données du S&P 500
sp500 = yf.download('^GSPC', start='2012-01-01')

# Calculer la SMA 200
sp500['SMA200'] = sp500['Close'].rolling(window=200).mean()

# Afficher les 5 dernières lignes pour vérifier
print(sp500.tail())

# Tracer les données
plt.figure(figsize=(14, 7))
plt.plot(sp500.index, sp500['Close'], label='S&P 500 Close')
plt.plot(sp500.index, sp500['SMA200'], label='SMA 200', linewidth=2)
plt.title('S&P 500 + SMA 200')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```
